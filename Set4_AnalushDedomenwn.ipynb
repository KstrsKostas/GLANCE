## Gaussian Distribution

```python
import numpy as np
import random
```

## n=5

```python
sigma=1
T1 = RealDistribution('gaussian', sigma)
S1=[T1.get_random_element() for _ in range(1000)]
```

```python
n=5
Md=[]
Sd=[]
S2d=[]
for j in range (10^5):
    M=0
    S=0
    S2=0
    for i in range(n):
        k=random.choice(S1)
        M+=(1/n)*k
        S+=(1/n)*(k-M)^2
        S2+=(1/(n-1))*(k-M)^2
    Md.append(M)
    Sd.append(S)
    S2d.append(S2)  
```

```python
Mhist=histogram(Md,bins=70, alpha=0.5,edgecolor='red',
                fill=True, linestyle='dashed', hatch='s',
                linewidth=1,fontsize=15,title=r' M, 10000 groups of n=5')

Shist=histogram(Sd,bins=70, alpha=0.5,edgecolor='red',
                     fill=True, linestyle='dashed', hatch='s',
                     linewidth=1,fontsize=15,title=r'S, 10000 groups of n=5')

S2hist=histogram(S2d,bins=70, alpha=0.5,edgecolor='red',
                     fill=True, linestyle='dashed', hatch='s',
                     linewidth=1,fontsize=15,title=r'S^2, 10000 groups of n=5')
```

```python
show(Mhist)
```

```python
show(Shist)
```

```python
show(S2hist)
```
