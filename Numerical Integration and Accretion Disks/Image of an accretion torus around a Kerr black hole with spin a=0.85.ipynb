## Hamiltonian Formulation:

### Metric of a stationary and axisymmetric spacetime:

 \begin{equation}
ds^2=g_{tt} dt^2 +g_{rr}dr^2 + g_{\theta\theta}d\theta^2+ g_{t\phi}dt d\phi +g_{\phi\phi}d\phi^2.
\end{equation}


The two primary Killing vector fields correspond to two invariant quantities, the energy $E$ and the angular momentum $L$, both normalized per unit mass,

\begin{equation}
E=-\xi^\alpha u_{\alpha}= -\left(g_{tt} \frac{dt}{d\lambda} + g_{t\phi}\frac{d\phi}{d\lambda}\right),
\label{eq:9}
\end{equation}

\begin{equation}
L=\eta^\alpha u_{\alpha}=g_{t\phi}\frac{dt}{d\lambda}+ g_{\phi\phi}\frac{d\phi}{d\lambda},
\label{eq:10}
\end{equation}

Using the Lagrangian:

\begin{equation}
\mathcal{L}=\frac{1}{2}g_{\alpha b} \dot{x}^\alpha \dot{x}^b,
\end{equation}

we can establish the equations governing geodesic motion in this spacetime. The associated generalized momenta are:

\begin{equation}
p_{\alpha}=\frac{\partial \mathcal{L}}{\partial \dot{x}^\alpha},
\end{equation}

and the Hamiltonian is:


\begin{equation}
\begin{split}
\mathcal{H}&=\frac{1}{m} \sum p_{\alpha} \dot{x}^\alpha - \mathcal{L} = -E \dot{t}+L \dot{\phi} + g_{rr}\dot{r}^2 +g_{\theta\theta} \dot{\theta}^2 -\mathcal{L} \\
&= \frac{1}{2}\left(g_{rr}\dot{r}^2 +g_{\theta\theta}\dot{\theta}^2 - \frac{L^2 g_{tt} + 2EL g_{t\phi} + E^2 g_{\phi\phi}}{\mathcal{D}}\right) \\
&=\frac{1}{2}\left(\frac{p_{r}^2}{g_{rr}}+\frac{p_{\theta}^2}{g_{\theta\theta}}-\frac{L^2 g_{tt} + 2EL g_{t\phi} + E^2 g_{\phi\phi}}{\mathcal{D}}\right),
\end{split}
\label{eq:hamiltonian}
\end{equation}

where $\mathcal{D}=g_{t\phi}^2 - g_{tt}g_{\phi\phi}$, is the Killing horizon equation of the spacetime

Hamilton's canonical equations yield the equations of motion:

\begin{equation}
\dot{x}^\alpha=\frac{\partial \mathcal{H}}{\partial p_\alpha},
\quad 
\dot{p}_\alpha=-\frac{\partial \mathcal{H}}{\partial x^\alpha}.
\end{equation}

 For photons, these are chosen to ensure the condition $\mathcal{H}=0$ but in general, the specification of initial positions depends on the problem's distinct details. For momenta, the relations are $p_t(0)=-E$ and $p_{\phi}(0)=L$. The full system of equations becomes:

\begin{equation}
\begin{aligned}
\dot{r}&=\frac{p_r}{g_{rr}} , &\dot{p_r}&=-\frac{\partial \mathcal{H}}{\partial r}, \\
\dot{\theta}&=\frac{p_\theta}{g_{\theta\theta}} , &\dot{p_\theta}&=-\frac{\partial \mathcal{H}}{\partial \theta}, \\
\dot{t}&=\frac{E g_{\phi\phi}+L g_{t\phi}}{\mathcal{D}} , &\dot{p_t}&=0, \\
\dot{\phi}&=-\frac{L g_{tt} + E g_{t\phi}}{\mathcal{D}} , &\dot{p_\phi}&=0.
\end{aligned}
\end{equation}

### Metric Definition (Kerr BH)

The Kerr black hole (BH) is often seen as the key black hole solution to Einstein's equations. In \emph{Boyer-Lindquist} coordinates $(t,r,\theta,\phi)$, it can describe the exterior spacetime geometry of a rotating BH through the line element:

\begin{equation}
\begin{split}
ds^2=g_{\alpha\beta}dx^\alpha dx^\beta = -\left(1-\frac{2mr}{\rho^2}\right)dt^2 - \frac{4a mr \sin^2\theta}{\rho^2}dt d\phi + \frac{\rho^2}{\Delta}dr^2 + \rho^2 d\theta^2 \\ + \left(r^2 + a^2 +\frac{2 a^2 m r \sin^2\theta}{\rho^2}\right)\sin^2\theta d\phi^2,
\end{split}
\label{eq:2.41}
\end{equation}

where $\rho^2 = r^2 + a^2 \cos^2\theta$ and 
$\Delta = r^2 -2mr + a^2 = \left(r-r_-\right)\left(r-r_+\right)$. It admits two event horizons at $r_{\pm}= M\pm \sqrt{M^2-a^2}$ and two \textit{ergosurfaces} at $R_{\pm}= M\pm \sqrt{M^2-a^2\cos^2\theta}$. 

\begin{equation}
E = -\boldsymbol{\xi} \cdot p_t \quad \text{and} \quad L = \boldsymbol{\eta} \cdot p_\phi,
\end{equation}

where $p$ is the particle's four-momentum. Here, $E$ and $L$ are the conserved energy and conserved angular momentum, respectively. While there's an evident integral from the four-velocity normalization $u^\alpha u_\alpha=-1,0$ (massive,massless), a fourth one was identified by B.Carter using the Hamilton-Jacobi method. The resultant four motion integrals are:

\begin{equation}
p_t=-E,  p_\phi = L_z,
\end{equation} 
\begin{equation}
{p_\theta}^2 + \cos^2\theta \left(a^2\left(\mu^2-E^2\right)+\frac{L^2_z}{\sin^2\theta}\right)=\mathcal{C},
\end{equation}
\begin{equation}
g^{\alpha\beta}p_\alpha p_\beta = -\mu^2,
\end{equation}

where $\mathcal{C}$ is called the Carter Constant that renders the system fully integrable. Using the inverse Kerr metric, we get:

\begin{equation}
g^{\alpha\beta}p_\alpha p_\beta= \frac{\Delta}{\rho^2}p_r ^2 + \frac{1}{\rho^2}p_\theta ^2 + \frac{\Delta - a^2 \sin^2\theta}{\rho^2 \Delta \sin^2 \theta}p_\phi ^2 - \frac{4M a r}{\rho^2 \Delta}p_\phi  p_t - \frac{\Sigma}{\rho^2 \Delta}p_t ^2.
\end{equation}

Substituting the conserved energy and angular momentum in the above equation, we derive:

\begin{equation}
\frac{\Delta}{\rho^2}p_r ^2 + \frac{1}{\rho^2}p_\theta ^2 + \frac{\Delta - a^2 \sin^2\theta}{\rho^2 \Delta \sin^2 \theta}L_z ^2 + \frac{4M a r}{\rho^2 \Delta}L_z E - \frac{\Sigma}{\rho^2 \Delta}E ^2= -\mu^2,
\end{equation}

and the full system of first order equations becomes:

\begin{equation}
\rho^2 \frac{d\theta}{d\lambda}=\sqrt{\Theta},
\label{eq:2.59}
\end{equation}
\begin{equation}
\rho^2 \frac{dr}{d\lambda}=\sqrt{R},
\label{eq:2.60}
\end{equation}
\begin{equation}
\rho^2 \frac{d\phi}{d\lambda}=-\left(aE-\frac{L_z}{\sin^2\theta}\right)+\frac{a}{\Delta}P,
\end{equation}
\begin{equation}
\rho^2 \frac{dt}{d\lambda}=-a\left(a E \sin^2\theta -L_z\right)+\frac{r^2 + a^2}{\Delta}P,
\label{eq:2.62}
\end{equation}
where
\begin{equation}
\Theta= \mathcal{C}-\cos^2\theta\left(a^2\left(\mu^2-E^2 \right)+\frac{L_z ^2}{\sin^2 \theta}\right),
\end{equation}
\begin{equation}
R=P^2-\Delta\left(\mu^2 r^2 + \left(L_z -a E\right)^2 + \mathcal{C}\right),
\end{equation}
with
\begin{equation}
P=E\left(r^2+a^2\right)-L_z a.
\end{equation}

To determine $r(\lambda)$ and $\theta(\lambda)$, we utilize quadrature from the initial two equations. This also aids in determining $\phi(\lambda)$ and $t(\lambda)$.

## Kerr Metric definition

```python
from sympy import symbols, diff, cos, sin, simplify, log
from sympy import lambdify
import numpy as np
## Kerr

m, a = symbols('m a')  
t, r, th, phi = symbols('t r th phi')  
pr, pth, pt, pphi = symbols('pr pth pt pphi')  


Sigma = simplify(r**2 + a**2 * cos(th)**2)
Delta = simplify(r**2 - 2*m*r + a**2)
g_tt = simplify(-(1 - 2*m*r/Sigma))
g_tphi = simplify(-2*m*a*r*sin(th)**2/Sigma)
g_phiphi = simplify((r**2 + a**2 + 2*m*a**2*r*sin(th)**2/Sigma)*sin(th)**2)
g_rr = simplify(Sigma/Delta)
g_thth = simplify(Sigma)
D = simplify(g_tphi**2 - g_tt * g_phiphi)
```

```python
E = -pt
L = pphi
H = ((1/2) * (pr**2/g_rr + pth**2/g_thth - (L**2 * g_tt + 2*E*L * g_tphi + E**2 * g_phiphi)/D))
print(f"Hamiltonian: {H}")
```

Create a function that takes the state vector as an argument, unpacks it and calculates the derivatives:

```python
def eom_system(λ, y, metric_params):
    t, r, th, φ, pr, pth, pt, pphi = y  


    params = metric_params.values()

    dt_dλ = tdot_func(r, th, pt, pphi, pr, pth, *params)
    dr_dλ = rdot_func(r, th, pt, pphi, pr, pth, *params)
    dth_dλ = thdot_func(r, th, pt, pphi, pr, pth,*params)
    dφ_dλ = phidot_func(r, th, pt, pphi, pr, pth,*params)
    dpr_dλ = prdot_func(r, th, pt, pphi, pr, pth,*params)
    dpth_dλ = pthdot_func(r, th, pt, pphi, pr, pth,*params)
    dpt_dλ = 0  
    dpphi_dλ = 0  

    return np.array([dt_dλ, dr_dλ, dth_dλ, dφ_dλ, dpr_dλ, dpth_dλ, dpt_dλ, dpphi_dλ])
```

We create the 'intersections' function that takes the numerical solution of the geodesic as an argument and finds its intersections with a torus with dimensions of our choice:

```python
def intersections(curve):
    curve = np.array(curve)

    # Calculate the x, y, z coordinates of each point on the geodesic
    r = curve[:, 2]
    theta = curve[:, 3]
    phi = curve[:, 4]
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

    # Filter out points that are not within the ring-like structure
    mask = (8.5 < r) & (r < 20) & (-3 < z) & (z < 3)

    if not np.any(mask):
        return [False, []]

    # Find segments of the curve that are within the disk
    segments = []
    in_segment = False
    for i in range(len(mask)):
        if mask[i] and not in_segment:  # start of a new segment
            in_segment = True
            start_idx = i
        elif not mask[i] and in_segment:  # end of a segment
            in_segment = False
            end_idx = i
            segments.append((start_idx, end_idx))

    # if the geodesic is still in a segment at the end, add that segment
    if in_segment:
        segments.append((start_idx, len(mask)-1))

    return [True, segments]
```

### calculate_total_angle_change: tracks and sums the angle between two successive points of the geodesic

```python
def calculate_total_angle_change(data_np, radius):
    # Extract r, theta, and phi values
    r_values = data_np[:, 2]
    theta_values = data_np[:, 3]
    phi_values = data_np[:, 4]

    # Filter points inside the sphere of interest
    inside_sphere_indices = np.where(r_values <= radius)[0]
    inside_sphere_data = data_np[inside_sphere_indices]

    # Convert to Cartesian coordinates
    x, y, z = spherical_to_cartesian2(inside_sphere_data[:, 2],
                                     inside_sphere_data[:, 3],
                                     inside_sphere_data[:, 4])

    # Calculate the total angle change
    total_angle = 0
    for i in range(1, len(x)):
        u = np.array([x[i-1], y[i-1], z[i-1]])
        v = np.array([x[i], y[i], z[i]])

        # Angle between u and v
        angle = np.arccos(np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v)))
        total_angle += angle

    return total_angle
```

# Radiative Transfer Formulation

The Lorentz Invariant intensity $\mathcal{I}$ of every light ray can be calculated through the radiative transfer equation ,
\begin{equation}
    \frac{d\mathcal{I}}{d\lambda}=\gamma^{-1}\left(\frac{j_0}{\nu^3_{0}}\right),
\end{equation}
where $\gamma=\nu/\nu_0$, with $\nu$ being the frequency and $"0"$ denotes quantities in the local rest frame. We have assumed zero absorption. At every point of the geodesic, this is transformed to the specific intensity using
\begin{equation}
\begin{split}
    dI&=d\mathcal{I}\nu_{obs}^3=\gamma^{-1}\left(\frac{j_0}{\nu_0^3}\right)\nu^3_{obs}d\lambda\\
    &=\gamma^2 j_0 d\lambda,
\end{split}
\end{equation}
which is the contribution to the intensity from a differential segment of the geodesic $d\lambda$. This transformation has to take place just before outputting the results, since the intensity is affected by the act of going from the emission to the observer's frame. We use the specific intensity at every point $I_\nu$, paired with its respective $\nu$, to calculate direct intensity images through
\begin{equation}
    dI=I_\nu d\nu.
\end{equation}

### $(r,\theta)$ function for the density/emission model of the torus 

```python
def spherical_to_cartesian(r, theta, phi):
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)
    return x, y, z



def density_model(r, theta):
    r0 = 0.8  # scale length
    r_total = np.sqrt((r*sin(theta)-12)**2 + (r*cos(theta))**2)  
    return np.exp(-r_total/r0)
```

### torus' velocity profile

```python
def calculate_azimuthal_velocity(r, theta, a_value, m_value):

    #omega = np.sqrt(m_value / (r**3 + a_value * np.sqrt(m_value * r)))  # omega = dphi/dt
    
    en=0.232
    rkappa=12    
    omega=(sqrt(m_value)/((r*np.sin(theta))^(3/2)+a_value*sqrt(m_value)))*(rkappa/(r*np.sin(theta)))^en
    
    

    vphi_fluid = omega

    return vphi_fluid
```

### calculate the intensity and energy shift / point 

```python
def intensity_at_point(points_4_velocity, E0, a_value, m_value,q_value):    
    λ,t, r, theta, phi, vt, vr, vtheta, vphi = points_4_velocity
    z = r * np.cos(theta)    
    boundaries=20, 8, 3, -3
    #density = density_model(r,z)
    density=density_model(r,theta)
    #density=1/r^2
    

    #metric definition (This one is JP)
    Σ = r**2 + a_value^2*cos(theta)^2
    Δ = r**2 - 2*m_value*r + a_value^2
    h = q_value*((m_value^3*r)/Σ^2)
    
    
    g = np.array([
        [(-(1 - (2*m_value*r)/Σ))*(1 + h), 0, 0, (-((2*m_value*a_value*r)/Σ))*sin(theta)^2*(1 + h)],
        [0, (Σ*(1 + h))/(Δ + h*a_value^2*sin(theta)^2), 0, 0],
        [0, 0, Σ, 0],
        [(-((2*m_value*a_value*r)/Σ))*sin(theta)^2*(1 + h), 0, 0, (r^2 + a_value^2 + (2*m_value*a_value^2*r*sin(theta)^2)/Σ + h*((a_value^2*(Σ + 2*m_value*r))/Σ)*sin(theta)^2)*sin(theta)^2]
    ])    
    
    #4-momentum of the photon
    
    k_0 = np.dot(g, np.array([vt, vr, vtheta, vphi]))

    
    vphi_fluid = calculate_azimuthal_velocity(r, theta, a_value, m_value)
    
    u_0_candidate = np.array([1, 0, 0, vphi_fluid]) 

    dot_product = np.dot(u_0_candidate, np.dot(g, u_0_candidate))
    
    vt_fluid = np.sqrt(-1 / dot_product)
    
    vphi_fluid = vphi_fluid*np.sqrt(-1 / dot_product)
    
    
    u_0 = np.array([vt_fluid, 0, 0, vphi_fluid])
    
    
    #E_0->1 our reference energy in the local rest frame
    E_0=1
    
    
    #"interaction" of the photon with the disk encoded in:
    k_0_u_0 = np.dot(k_0, u_0)
    
    
    E_obs=-1/k_0_u_0
    
    gamma_inv=1/E_obs


    optical_depth = 0
    intensity = 0

    alpha0 = 0  # Here, we use density as a proxy for the local rest-frame absorption coefficient

    optical_depth += gamma_inv * alpha0

    j0 = density  
    intensity = gamma_inv * (j0 / E_0**3)*E_obs**3

    ratio=E_obs/E_0

    

    return intensity, ratio
```

### parallelization

```python
def input_data_generator(intersecting_geodesics_dict):
    for key, geodesic in intersecting_geodesics_dict.items():
        (b_al_pair, inters,color) = key  
        yield (b_al_pair, geodesic, inters, color)


def process_geodesic(input_data):
    key, geodesic_data, inters,color = input_data
    geodesic_4_velocity = calc_4_velocity4(geodesic_data)
    intensities = []
    ratios = []   
    for i in range(1, len(geodesic_4_velocity) - 1):        
        main_point = np.array([geodesic_4_velocity[i][0]] + geodesic_4_velocity[i][1][:8])
        main_point = main_point.tolist()
        
        intensity, ratio = intensity_at_point(main_point, 1, 0.85, 1, 1)
        if intensity is None or ratio is None:
            continue
        
        intensities.append(intensity)
        ratios.append(ratio)

        if i == len(geodesic_4_velocity) - 2:
            last_point_data = (intensity, ratio, main_point[2:5])

    return (key,intensities,ratios, last_point_data, inters,color)


#"key: it now is just the (b,al)-impact parameter pair i.e. a single geodesic
#"intensities: list of intensities for every point of the geodesic'
#"ratios: list of ratios for every point of the geodesic'
#"last point_data: (intensity,ratio,coordinates) of the last point of the geodesic----
#"------> useful for optically thick cases i.e. we consider only the boundary of the torus
#"color: mathematical shadows
```
