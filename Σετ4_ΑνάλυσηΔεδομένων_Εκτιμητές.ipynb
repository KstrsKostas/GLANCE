## Gaussian Distribution

```python
import numpy as np
import random
```

Αρχικά κατασκευάζουμε μια τυποποιημένη κανονική κατανομή ($μ=0$ και $σ=1$) με 10000 τυχαία στοιχεία.

```python
sigma=1
T1 = RealDistribution('gaussian', sigma)
S1=[T1.get_random_element() for _ in range(1000)]
```

Σκοπός μας είναι από το παραπάνω τυχαίο δείγμα να διαλέξουμε $n$ παρατηρήσεις $10^5$ φορές και να υπολογίσουμε για κάθε μία απο αυτές τις φορές τους τρεις ζητούμενους εκτιμητές. Οι εκτιμητές αυτοί θα ακολουθούν μια τυχαία κατανομή της οποίας και θα υπολογίζουμε κάθε φορά την μέση τιμή και την τυπική απόκλιση ώστε να μπορέσουμε να βγάλουμε στο τέλος τα συμπεράσματα μας.  

## n=5 



```python
n=5
Md=[]
Sd=[]
S2d=[]
for j in range (10^5):
    M=0
    S=0
    S2=0
    for i in range(n):
        k=random.choice(S1)
        M+=(1/n)*k
        S+=(1/n)*(k-M)^2
        S2+=(1/(n-1))*(k-M)^2
    Md.append(M)
    Sd.append(S)
    S2d.append(S2)  
```

```python
Mhist=histogram(Md,bins=70, alpha=0.5,edgecolor='red',
                fill=True, linestyle='dashed', hatch='s',
                linewidth=1,fontsize=15,title=r' M, 10000 groups of n=5')

Shist=histogram(Sd,bins=70, alpha=0.5,edgecolor='red',
                     fill=True, linestyle='dashed', hatch='s',
                     linewidth=1,fontsize=15,title=r'S, 10000 groups of n=5')

S2hist=histogram(S2d,bins=70, alpha=0.5,edgecolor='red',
                     fill=True, linestyle='dashed', hatch='s',
                     linewidth=1,fontsize=15,title=r'S^2, 10000 groups of n=5')
```

```python
show(Mhist)
```

```python
show(Shist)
```

```python
show(S2hist)
```

```python
mean_M1=np.mean(Md)
mean_S11=np.mean(Sd)
mean_S21=np.mean(S2d)
```

### Τυπική Απόκλιση

```python
std_M1=np.std(Md)
std_S11=np.std(Sd)
std_S21=np.std(S2d)
```

### Bias

```python
b_M1=np.mean(Md)-0
b_S11=np.mean(Sd)-1
b_S21=np.mean(S2d)-1
```

# Συμπεράσματα

Επιλέξαμε να μελετήσουμε τις περιπτώσεις $n=5$, $n=10$, $n=100$ και $n=1000$. Βλέπουμε στον παραπάνω πίνακα πως με την άυξηση του αριθμού των δειγμάτων $n$, οι εκτιμητές προσεγγίζουν τις αναμενόμενες τιμές $μ=0$ και $σ=1$.  Παρατηρούμε πως τουλάχιστον για μικρά $n$ ο εκτιμητής $M$ δεν είναι αμερόληπτος. Έχουμε απόκλιση από την αναμενόμενη μέση τιμή στο 3ο δεκαδικό μέχρι και για $n=100$ κάτι το οποίο όμως δεν ισχύει και για την τυπική απόκλιση η οποία συμπεριφέρεται όπως θέλουμε για να έχουμε έναν καλό εκτιμητή. Οι εκτιμητές $S$ και $S^2$ είναι αμερόληπτοι το οποίο φαίνεται απο τις τιμές των $b_{S^2}$ και $b_{S^2}$ που τείνουν στο $0$. Ο εκτιμητής $S*^2$ είναι ο καλύτερος των τριών δεδομένου οτι προσεγγίζει γρηγορότερα την αναμενόμενη τυπική απόκλιση $σ=1$.
